```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Fieldfox_plot(file):
    filename = os.path.splitext(os.path.basename(file))[0]
    df = pd.read_csv(file,header=0,engine='python',skiprows=42)
    df.columns = ['Frequency [Hz]', 'Clear-Write [dBm]', 'Max [dBm]', 'Min [dBm]', 'Average [dBm]']
    df['Frequency [Hz]'] = pd.to_numeric(df['Frequency [Hz]'], errors='coerce')
    print(df)

    plt.figure(figsize=(10, 5))
    #sometimes skiprows will need to change from 41 to 44 because of the FPH configuration on data colect
    plt.plot(df['Frequency [Hz]']*10e-7,df['Clear-Write [dBm]'], label = 'Clear-Write')
    plt.plot(df['Frequency [Hz]']*10e-7,df['Max [dBm]'], label = 'Max')
    plt.plot(df['Frequency [Hz]']*10e-7,df['Min [dBm]'], label = 'Min')
    plt.plot(df['Frequency [Hz]']*10e-7,df['Average [dBm]'], label = 'Average')


    plt.legend(loc='lower right', fontsize = 30)
    plt.xlabel('Frequency (MHz)')
    plt.ylabel('Power (dBm)')
    plt.title(filename)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def FPH_plot(file):
    filename = os.path.splitext(os.path.basename(file))[0]
    df = pd.read_csv(file,header=0,engine='python',skiprows=42)
    print(df)

    plt.figure(figsize=(10,5))
    #sometimes skiprows will need to change from 41 to 44 because of the FPH configuration on data colect
    plt.plot(df['Frequency [Hz]']/1e6,df['Magnitude [dBm]'], label = 'Max')
    plt.plot(df['Frequency [Hz]']/1e6,df['Magnitude [dBm].1'], label = 'Average')
    plt.legend(loc='upper right', fontsize = 10)
    plt.xlabel('Hz', size = 10)
    plt.ylabel('dBm', size = 10)
    plt.tick_params(axis='both', labelsize = 10)
    #plt.title(names_[i], size = 30)
    plt.title(filename)
    plt.ylim(-120,-70)
    plt.grid()
    plt.show()
```
